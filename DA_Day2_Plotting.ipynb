Problem-2:Plot the scores of 2 students in  5 different subjects as x axis and marks as y axis

```python
St1=[86,48,96,98,85]
St2=[49,38,57,38,98]
sub=["Maths","Science","Telugu","English","Social"]
plt.plot(sub,St1,color="b",label="student1")
plt.plot(sub,St2,color="r",label="student2")
plt.legend()
plt.title("Marks vs Students")
```

```python
St1=[86,48,96,98,85]
St2=[49,38,57,38,98]
sub=["Maths","Science","Telugu","English","Social"]
plt.subplot(2,1,1) #subplot(r,c,position)
plt.plot(sub,St1,color="b",label="student1")
plt.legend()
plt.title("student1")
print("\n")
plt.subplot(2,1,2)
plt.plot(sub,St2,color="r",label="student2")
plt.title("student2")
plt.legend()
```

Create subplots(bar) showing the profits of two companies
Given:Revenue of company A and B in lakhs

```python
A=np.array([230,560,780,127,128])
B=np.array([200,160,270,127,400])
print("Revenues of Company A",A)
print("Revenues of Company B",B)
profA=np.diff(A)
print("Profits of A",profA)
profB=np.diff(B)
print("Profits of B",profB)
X=["19-20","20-21","21-22","22-23"]
plt.subplot(1,2,1)
plt.bar(X,profA,label="profit A",color="k")
plt.legend(loc="lower left")
plt.subplot(1,2,2)
plt.bar(X,profB,label="profit B",color="g")
plt.legend(loc="lower left")
```
